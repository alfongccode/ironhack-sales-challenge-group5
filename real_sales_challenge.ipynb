```python
import pandas as pd
```

```python
real_sales = pd.read_csv('./REAL_DATA.csv')
display(real_sales.head(5))
display(real_sales.sample(5))
display(real_sales.tail(5))

display(real_sales.isna().sum())
display(real_sales.dtypes)
display(real_sales['state_holiday'].unique())
display(real_sales.describe())
display(real_sales.info())
days_of_week = real_sales['day_of_week'].unique()
days_of_week.sort()
display(days_of_week)
```

```python
real_sales_clean = real_sales.drop(columns=['index', 'store_ID', 'date'])
real_sales_fe = pd.get_dummies(real_sales_clean, columns=['state_holiday'], drop_first=True, dtype=int)
display(real_sales_fe)
```

```python
import pickle

with open("sales_model.pkl", "rb") as file:
    model = pickle.load(file)
    
real_sales_predict = model.predict(real_sales_fe)
display(real_sales_predict)

real_sales['sales'] = real_sales_predict
display(real_sales)

display(real_sales.describe())

real_sales.to_csv('group5.csv', index=False, encoding='utf-8')
```
